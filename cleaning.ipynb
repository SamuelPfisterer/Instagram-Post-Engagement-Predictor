Here we clean the final file (with engineered features), before we encode the relevant columns and feed it to the models

```python
# Required imports
import pandas as pd
```

```python
# Loading the feature engineered file
input_file = 'feature_engineered_metadata.csv'
df = pd.read_csv(input_file)
```

# Dropping "distracting" Rows

```python
# Filter for all the rows where the post_type is not video and is_private = false, and comment_disabled = false
filtered_df = df[(df['is_video'] == False) & (df['is_private'] == False) & (df['is_ad'] == False) & (df['comments_disabled'] == False)]
```

# Dropping "irrelevant" Columns

```python
#  Dropping columns that we won't use in our ML models

# We define all the columns that we want to drop
columns_to_drop = ['file_name', 'full_name', 'is_verified', 'owner_id',
                   'is_private', 'is_ad', 'post_id',
                   'comment_count', 'timestamp', 'comments_disabled', 'is_video',
                   'caption_text', 'country_code', 'city_name', 'location_name',
                   'Date-Time']

# Now we drop them from the filtered_df
filtered_df = filtered_df.drop(columns=columns_to_drop)
```

# Storing the Cleaned Metadata

```python
# Storing the cleaned metadata in the cleaned_metadata file
output_file = 'cleaned_metadata.csv'
filtered_df.to_csv(output_file, index=False)
```
